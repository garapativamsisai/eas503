```python
import pandas as pd
import requests
from io import BytesIO

def fetch_raw_data(year: int, month: int) -> pd.DataFrame:
    url = f"https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_{year}-{month:02}.parquet"
    
    response = requests.get(url)
    if response.status_code == 200:
        df = pd.read_parquet(BytesIO(response.content), engine="pyarrow")
        print(f"Successfully fetched data for {year}-{month:02}, shape: {df.shape}")
        return df
    else:
        raise Exception(f"{url} is not available")
```

```python
df = fetch_raw_data(2024, 1) 
```
